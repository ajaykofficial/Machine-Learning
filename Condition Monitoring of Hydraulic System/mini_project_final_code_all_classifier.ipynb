```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.feature_selection import RFECV

# Load your dataset into 'df'
df = pd.read_csv(r"C:\Users\ajayk\Desktop\Machine Learning\Machine-Learning\Condition Monitoring of Hydraulic System\condition_monitoring_dataset.csv")

# Define your features (X) and target variables (y)
features = ['PS1', 'PS2', 'PS3', 'PS4', 'PS5', 'PS6', 'FS1', 'FS2', 'TS1', 'TS2', 'TS3', 'TS4', 'P1', 'VS1', 'CE1', 'CP1', 'SE1']
target_variables = ['y_coolerCondition', 'y_valveCondition', 'y_pumpLeak', 'y_hydraulicAcc']

# Define classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(probability=True),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Bagging': BaggingClassifier(),
    'Gradient Boosting': GradientBoostingClassifier()
}

def evaluate_classifier(y_true, y_pred, classifier_name, is_binary=False):
    print(f"Evaluating {classifier_name}...\n")
    print(f"Classification Report for {classifier_name}:\n")
    print(classification_report(y_true, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)

    # Heatmap for Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=np.unique(y_true),
        yticklabels=np.unique(y_true),
    )
    plt.title(f'Confusion Matrix for {classifier_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    if is_binary:
        # Plot ROC curve for binary classification
        fpr, tpr, _ = roc_curve(y_true, y_pred)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {classifier_name}')
        plt.legend(loc="lower right")
        plt.show()

    else:
        # Multiclass ROC curve
        n_classes = len(np.unique(y_true))
        fpr = dict()
        tpr = dict()
        roc_auc = dict()

        for i in range(n_classes):
            fpr[i], tpr[i], _ = roc_curve(label_binarize(y_true, classes=np.unique(y_true))[:, i],
                                           label_binarize(y_pred, classes=np.unique(y_true))[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])

        # Plot ROC curve for each class
        plt.figure(figsize=(8, 6))
        colors = ['darkorange', 'green', 'blue', 'purple'][:n_classes]  # Add more colors if needed
        for i, color in zip(range(n_classes), colors):
            plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {classifier_name}')
        plt.legend(loc="lower right")
        plt.show()

    print("-" * 50)

# Create an empty DataFrame to store the result summary
result_summary = pd.DataFrame(columns=['Classifier', 'Target Variable', 'Selected Features', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'])

# Loop through classifiers and target variables
for classifier_name, classifier in classifiers.items():
    for target_variable in target_variables:
        print(f"Evaluating {classifier_name} for {target_variable}...\n")

        # Extract target variable (y) for the current target
        y = df[target_variable]

        # Extract features (X) once outside the loop
        X_all = df[features]

        # Perform RFECV for the current target variable
        rf_classifier = RandomForestClassifier(random_state=42)
        rfecv = RFECV(estimator=rf_classifier, step=1, cv=5)
        X_rfecv = rfecv.fit_transform(X_all, y)
        selected_features = [feature for feature, selected in zip(features, rfecv.support_) if selected]
        print(f"Selected Features: {selected_features}\n")

        # Split the data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(df[selected_features], y, test_size=0.2, random_state=42)

        # Train the classifier on the specific target variable
        classifier.fit(X_train, y_train)

        # Make predictions
        y_pred = classifier.predict(X_test)

        # Check if binary or multiclass and update the function call
        is_binary = len(np.unique(y)) == 2
        evaluate_classifier(y_test, y_pred, classifier_name, is_binary)

        # Compute and append results to the summary DataFrame
        report = classification_report(y_test, y_pred, output_dict=True)
        accuracy = report['accuracy']
        precision = report['weighted avg']['precision']
        recall = report['weighted avg']['recall']
        f1_score = report['weighted avg']['f1-score']

        # Multiclass ROC-AUC
        if not is_binary:
            n_classes = len(np.unique(y_test))
            fpr = dict()
            tpr = dict()
            roc_auc = dict()

            for i in range(n_classes):
                y_true_binary = label_binarize(y_test, classes=np.unique(y_test))
                y_pred_binary = label_binarize(y_pred, classes=np.unique(y_test))

                fpr[i], tpr[i], _ = roc_curve(y_true_binary[:, i], y_pred_binary[:, i])
                roc_auc[i] = auc(fpr[i], tpr[i])

            # Average ROC-AUC over all classes
            avg_roc_auc = np.mean(list(roc_auc.values()))

        result_summary = result_summary.append({
            'Classifier': classifier_name,
            'Target Variable': target_variable,
            'Selected Features': selected_features,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1_score,
            'AUC': avg_roc_auc if not is_binary else roc_auc
        }, ignore_index=True)

# Display the result summary
print("Result Summary:")
result_summary
```
