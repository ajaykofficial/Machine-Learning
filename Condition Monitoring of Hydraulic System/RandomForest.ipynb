# Importing Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics 
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# Loading Dataset

```python
df= pd.read_csv(r"C:\Users\ajayk\Desktop\Machine Learning\Machine-Learning\Condition Monitoring of Hydraulic System\condition_monitoring_of_Hydraulic_system.csv")
```

# Assigning values to X and y

```python
X = df[['PS1', 'PS2', 'PS3', 'PS4', 'PS5', 'PS6', 'FS1', 'FS2',
       'TS1', 'TS2', 'TS3', 'TS4', 'P1', 'VS1', 'CE1', 'CP1', 'SE1']]
y = df[['y_coolerCondition', 'y_valveCondition', 'y_pumpLeak', 'y_hydraulicAcc',
       'y_stableFlag']]
```

```python
X
```

```python
y
```

# Splitting dataset to train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
X_train
```

```python
y_train
```

# Initialize separate classifiers for each target variable

```python
nb_cooler = RandomForestClassifier(n_estimators=100)
nb_valve = RandomForestClassifier(n_estimators=100)
nb_pump = RandomForestClassifier(n_estimators=100)
nb_hydraulic = RandomForestClassifier(n_estimators=100)
nb_stable = RandomForestClassifier(n_estimators=100)
```

# Fit each classifier on the corresponding target variable

```python
nb_cooler.fit(X_train, y_train['y_coolerCondition'])
nb_valve.fit(X_train, y_train['y_valveCondition'])
nb_pump.fit(X_train, y_train['y_pumpLeak'])
nb_hydraulic.fit(X_train, y_train['y_hydraulicAcc'])
nb_stable.fit(X_train, y_train['y_stableFlag'])
```

# Make predictions on the test data for each classifier

```python
y_pred_cooler = nb_cooler.predict(X_test)
y_pred_valve = nb_valve.predict(X_test)
y_pred_pump = nb_pump.predict(X_test)
y_pred_hydraulic = nb_hydraulic.predict(X_test)
y_pred_stable = nb_stable.predict(X_test)
```

# Evaluate the performance for each classifier

```python
accuracy_cooler = metrics.accuracy_score(y_test['y_coolerCondition'], y_pred_cooler)
accuracy_valve = metrics.accuracy_score(y_test['y_valveCondition'], y_pred_valve)
accuracy_pump = metrics.accuracy_score(y_test['y_pumpLeak'], y_pred_pump)
accuracy_hydraulic = metrics.accuracy_score(y_test['y_hydraulicAcc'], y_pred_hydraulic)
accuracy_stable = metrics.accuracy_score(y_test['y_stableFlag'], y_pred_stable)
```

# Print results

```python
print("Accuracy - Cooler Condition:",accuracy_cooler)
print("Accuracy - Valve Condition:",accuracy_valve)
print("Accuracy - Pump Leak:",accuracy_pump)
print("Accuracy - Hydraulic Acc:",accuracy_hydraulic)
print("Accuracy - Stable Flag:",accuracy_stable)
```

# Generate classification reports and confusion matrices

```python
report_cooler = classification_report(y_test['y_coolerCondition'], y_pred_cooler)
matrix_cooler = confusion_matrix(y_test['y_coolerCondition'], y_pred_cooler)

report_valve = classification_report(y_test['y_valveCondition'], y_pred_valve)
matrix_valve = confusion_matrix(y_test['y_valveCondition'], y_pred_valve)

report_pump = classification_report(y_test['y_pumpLeak'], y_pred_pump)
matrix_pump = confusion_matrix(y_test['y_pumpLeak'], y_pred_pump)

report_hydraulic = classification_report(y_test['y_hydraulicAcc'], y_pred_hydraulic)
matrix_hydraulic = confusion_matrix(y_test['y_hydraulicAcc'], y_pred_hydraulic)

report_stable = classification_report(y_test['y_stableFlag'], y_pred_stable)
matrix_stable = confusion_matrix(y_test['y_stableFlag'], y_pred_stable)
```

# Print classification reports and confusion matrices

```python
print("Classification Report - Cooler Condition:")
print(report_cooler)
print("Confusion Matrix - Cooler Condition:")
print(matrix_cooler)
print("\n")

print("Classification Report - Valve Condition:")
print(report_valve)
print("Confusion Matrix - Valve Condition:")
print(matrix_valve)
print("\n")

print("Classification Report - Pump Leak:")
print(report_pump)
print("Confusion Matrix - Pump Leak:")
print(matrix_pump)
print("\n")

print("Classification Report - Hydraulic Acc:")
print(report_hydraulic)
print("Confusion Matrix - Hydraulic Acc:")
print(matrix_hydraulic)
print("\n")

print("Classification Report - Stable Flag:")
print(report_stable)
print("Confusion Matrix - Stable Flag:")
print(matrix_stable)
```

# Plotting confusion matrix as heatmap

```python
def plot_confusion_matrix_heatmap(matrix, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
plot_confusion_matrix_heatmap(matrix_cooler, title='Cooler Condition Confusion Matrix')
```

```python
plot_confusion_matrix_heatmap(matrix_valve, title='Valve Condition Confusion Matrix')
```

```python
plot_confusion_matrix_heatmap(matrix_hydraulic, title='Hydraulic Acc Confusion Matrix')
```

```python
plot_confusion_matrix_heatmap(matrix_pump, title='Pump Leak Confusion Matrix')
```

```python
plot_confusion_matrix_heatmap(matrix_stable, title='Stable Flag Confusion Matrix')
```
