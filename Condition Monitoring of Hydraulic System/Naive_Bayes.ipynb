# Importing modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import seaborn as sns
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# Loading datset

```python
df = pd.read_csv(r"C:\Users\ajayk\Desktop\Machine Learning\Machine-Learning\Condition Monitoring of Hydraulic System\condition_monitoring_of_Hydraulic_system.csv")
```

# Assigning values to X and y

```python
X = df[['PS1', 'PS2', 'PS3', 'PS4', 'PS5', 'PS6', 'FS1', 'FS2',
       'TS1', 'TS2', 'TS3', 'TS4', 'P1', 'VS1', 'CE1', 'CP1', 'SE1']]
y = df[['y_coolerCondition', 'y_valveCondition', 'y_pumpLeak', 'y_hydraulicAcc',
       'y_stableFlag']]
```

```python
X
```

```python
y
```

# Splitting dataset to train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

# Initialize separate classifiers for each target variable

```python
nb_cooler = MultinomialNB()
nb_valve = MultinomialNB()
nb_pump = MultinomialNB()
nb_hydraulic = MultinomialNB()
nb_stable = MultinomialNB()
```

# Fit each classifier on the corresponding target variable

```python
nb_cooler.fit(X_train, y_train['y_coolerCondition'])
nb_valve.fit(X_train, y_train['y_valveCondition'])
nb_pump.fit(X_train, y_train['y_pumpLeak'])
nb_hydraulic.fit(X_train, y_train['y_hydraulicAcc'])
nb_stable.fit(X_train, y_train['y_stableFlag'])
```

# Make predictions on the test data for each classifier

```python
y_pred_cooler = nb_cooler.predict(X_test)
y_pred_valve = nb_valve.predict(X_test)
y_pred_pump = nb_pump.predict(X_test)
y_pred_hydraulic = nb_hydraulic.predict(X_test)
y_pred_stable = nb_stable.predict(X_test)
```

# Evaluate the performance for each classifier

```python
accuracy_cooler = accuracy_score(y_test['y_coolerCondition'], y_pred_cooler)
accuracy_valve = accuracy_score(y_test['y_valveCondition'], y_pred_valve)
accuracy_pump = accuracy_score(y_test['y_pumpLeak'], y_pred_pump)
accuracy_hydraulic = accuracy_score(y_test['y_hydraulicAcc'], y_pred_hydraulic)
accuracy_stable = accuracy_score(y_test['y_stableFlag'], y_pred_stable)
```

# Print results

```python
print(f'Accuracy - Cooler Condition: {accuracy_cooler:.2f}')
print(f'Accuracy - Valve Condition: {accuracy_valve:.2f}')
print(f'Accuracy - Pump Leak: {accuracy_pump:.2f}')
print(f'Accuracy - Hydraulic Acc: {accuracy_hydraulic:.2f}')
print(f'Accuracy - Stable Flag: {accuracy_stable:.2f}')
```

```python
report_cooler = classification_report(y_test['y_coolerCondition'], y_pred_cooler)
matrix_cooler = confusion_matrix(y_test['y_coolerCondition'], y_pred_cooler)

report_valve = classification_report(y_test['y_valveCondition'], y_pred_valve)
matrix_valve = confusion_matrix(y_test['y_valveCondition'], y_pred_valve)

report_pump = classification_report(y_test['y_pumpLeak'], y_pred_pump)
matrix_pump = confusion_matrix(y_test['y_pumpLeak'], y_pred_pump)

report_hydraulic = classification_report(y_test['y_hydraulicAcc'], y_pred_hydraulic)
matrix_hydraulic = confusion_matrix(y_test['y_hydraulicAcc'], y_pred_hydraulic)

report_stable = classification_report(y_test['y_stableFlag'], y_pred_stable)
matrix_stable = confusion_matrix(y_test['y_stableFlag'], y_pred_stable)
```

# Function to plot confusion matrix as heatmap

```python
def plot_confusion_matrix_heatmap(matrix, class_names, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

# Function to plot ROC curve

```python
def plot_roc_curve(y_true, y_score, title):
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc='lower right')
    plt.show()
```

# Get probability scores for each class

```python
y_prob_cooler = nb_cooler.predict_proba(X_test)[:, 1]
y_prob_valve = nb_valve.predict_proba(X_test)[:, 1]
y_prob_pump = nb_pump.predict_proba(X_test)[:, 1]
y_prob_hydraulic = nb_hydraulic.predict_proba(X_test)[:, 1]
y_prob_stable = nb_stable.predict_proba(X_test)[:, 1]
```

# Plot ROC curves

```python
plot_roc_curve(y_test['y_coolerCondition'], y_prob_cooler, title='Cooler Condition ROC Curve')
plot_roc_curve(y_test['y_valveCondition'], y_prob_valve, title='Valve Condition ROC Curve')
plot_roc_curve(y_test['y_pumpLeak'], y_prob_pump, title='Pump Leak ROC Curve')
plot_roc_curve(y_test['y_hydraulicAcc'], y_prob_hydraulic, title='Hydraulic Acc ROC Curve')
plot_roc_curve(y_test['y_stableFlag'], y_prob_stable, title='Stable Flag ROC Curve')
```
