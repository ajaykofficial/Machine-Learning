```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

# Assigning values to X and y

```python
X = df[['PS1', 'PS2', 'PS3', 'PS4', 'PS5', 'PS6', 'FS1', 'FS2',
       'TS1', 'TS2', 'TS3', 'TS4', 'P1', 'VS1', 'CE1', 'CP1', 'SE1']]
y = df[['y_coolerCondition', 'y_valveCondition', 'y_pumpLeak', 'y_hydraulicAcc',
       'y_stableFlag']]
```

```python
X
```

```python
y
```

# Splitting dataset to train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

# Initialize separate classifiers for each target variable

```python
nb_cooler = MultinomialNB()
nb_valve = MultinomialNB()
nb_pump = MultinomialNB()
nb_hydraulic = MultinomialNB()
nb_stable = MultinomialNB()
```

# Fit each classifier on the corresponding target variable

```python
nb_cooler.fit(X_train, y_train['y_coolerCondition'])
nb_valve.fit(X_train, y_train['y_valveCondition'])
nb_pump.fit(X_train, y_train['y_pumpLeak'])
nb_hydraulic.fit(X_train, y_train['y_hydraulicAcc'])
nb_stable.fit(X_train, y_train['y_stableFlag'])
```

# Make predictions on the test data for each classifier

```python
y_pred_cooler = nb_cooler.predict(X_test)
y_pred_valve = nb_valve.predict(X_test)
y_pred_pump = nb_pump.predict(X_test)
y_pred_hydraulic = nb_hydraulic.predict(X_test)
y_pred_stable = nb_stable.predict(X_test)
```
